**Decorator** very easy peasy if you dont want to implemnt is yourself which i dont want. <br> so im going to use built in python decorator <br>

```python
from functools import wraps

def decorator_name(f):
    @wraps(f)
    def decorated(*args, **kwargs):
        if not can_run:
            return "Function will not run"
        return f(*args, **kwargs)
    return decorated

@decorator_name
def func():
    return("Function is running")

can_run = True
print(func())
# output: Function is running

can_run = False
print(func())
# output: Function will not run
```

```python
from functools import wraps
def user_auth(fn):
  def wrapper(request,*args,**kwrgs):
    if not request.get('user'):
        return {"error": "Authentication required"}, 401

    return fn(request,*args,**kwrgs)
  return wrapper

@user_auth
def get_user_profile(request):

    if request['user'].get('role') != 'user':
        return {"error": "You do not have permission to view this resource"}, 403

    user_profile = {"name": "Alice", "email": "alice@example.com"}
    return user_profile, 200
@user_auth
def delete_user(request):

    if request['user'].get('role') != 'admin':
        return {"error": "You do not have permission to delete this resource"}, 403

    # Logic to delete the user
    return {"message": "User deleted successfully"}, 200

request={"user": {"role": "admin"}}
print(delete_user(request))
```
