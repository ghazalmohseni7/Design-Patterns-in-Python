```python
# shallow copy
import copy

# Original object with nested list
original = [1, [2, 3], 4]

# Create a shallow copy
shallow_copied = copy.copy(original)

# Modify the nested list in the copied object
shallow_copied[1][0] = 'a'

print("Original:", original)  # Output: [1, ['a', 3], 4]
print("Shallow Copied:", shallow_copied)  # Output: [1, ['a', 3], 4]
```

```python
# deep copy
import copy

# Original object with nested list
original = [1, [2, 3], 4]

# Create a deep copy
deep_copied = copy.deepcopy(original)

# Modify the nested list in the copied object
deep_copied[1][0] = 'a'

print("Original:", original)  # Output: [1, [2, 3], 4]
print("Deep Copied:", deep_copied)  # Output: [1, ['a', 3], 4]
```
