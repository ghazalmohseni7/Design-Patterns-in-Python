```python
class EmailNotification:
    def send(self, message):
        print("Sending email notification...")
        print(f"Email: {message}")

class SMSNotification:
    def send(self, message):
        print("Sending SMS notification...")
        print(f"SMS: {message}")

class PushNotification:
    def send(self, message):
        print("Sending push notification...")
        print(f"Push: {message}")

class NotificationService:
    def send_notification(self, notification_type, message):
        if notification_type == "email":
            notification = EmailNotification()
        elif notification_type == "sms":
            notification = SMSNotification()
        elif notification_type == "push":
            notification = PushNotification()
        else:
            raise ValueError("Unknown notification type")

        notification.send(message)

# Client code
def main():
    service = NotificationService()
    service.send_notification("email", "Hello via Email!")
    service.send_notification("sms", "Hello via SMS!")
    service.send_notification("push", "Hello via Push!")
    service.send_notification("fax", "Hello via Fax!")  # This will raise an error

if __name__ == "__main__":
    main()
```

```python
from abc import ABC ,abstractmethod
class NotificationProductInterface(ABC):
  @abstractmethod
  def send(self, message):
    pass

class EmailNotificationProductImplementation(NotificationProductInterface):
  def send(self,message):
        print("Sending email notification...")
        print(f"Email: {message}")

class SMSNotificationProductImplementation(NotificationProductInterface):
  def send(self,message):
        print("Sending SMS notification...")
        print(f"SMS: {message}")

class PushNotificationProductImplementation(NotificationProductInterface):
  def send(self,message):
        print("Sending push notification...")
        print(f"Push: {message}")


class NotificationServiceCreatorInterface(ABC):
  @abstractmethod
  def create_notification(self)->NotificationProductInterface:
    pass
  def send_notification(self,message):
    notif=self.create_notification()
    notif.send(message)


class EmailNotificationServiceCreatorImplementation(NotificationServiceCreatorInterface):
  def create_notification(self):
    return EmailNotificationProductImplementation()
class SMSNotificationServiceCreatorImplementation(NotificationServiceCreatorInterface):
  def create_notification(self):
    return SMSNotificationProductImplementation()
class PushNotificationServiceCreatorImplementation(NotificationServiceCreatorInterface):
  def create_notification(self):
    return PushNotificationProductImplementation()

email_service = EmailNotificationServiceCreatorImplementation()
email_service.send_notification("Hello via Email!")

sms_service = SMSNotificationServiceCreatorImplementation()
sms_service.send_notification("Hello via SMS!")

push_service = PushNotificationServiceCreatorImplementation()
push_service.send_notification("Hello via Push!")
```

```python
from abc import ABC, abstractmethod
import sqlite3
import psycopg2
import mysql.connector

# Product Interface
class DataBaseProductInterface(ABC):
    @abstractmethod
    def connect(self):
        pass

# Concrete Products
class SqliteDataBaseProductImplementation(DataBaseProductInterface):
    def connect(self):
        connection = sqlite3.connect('example.db')
        print("Connected to SQLite")
        return connection

class PSQLDataBaseProductImplementation(DataBaseProductInterface):
    def connect(self):
        connection = psycopg2.connect(
            dbname="example",
            user="user",
            password="password",
            host="localhost"
        )
        print("Connected to PostgreSQL")
        return connection

class MysqlDataBaseProductImplementation(DataBaseProductInterface):
    def connect(self):
        connection = mysql.connector.connect(
            user="user",
            password="password",
            host="localhost",
            database="example"
        )
        print("Connected to MySQL")
        return connection

# Creator Interface
class DatabaseServiceCreatorInterface(ABC):
    @abstractmethod
    def create_db(self) -> DataBaseProductInterface:
        pass

    def create_connection(self):
        db = self.create_db()
        return db.connect()

# Concrete Creators
class SQLITEDatabaseServiceCreatorImplementation(DatabaseServiceCreatorInterface):
    def create_db(self) -> DataBaseProductInterface:
        return SqliteDataBaseProductImplementation()

class PSQLDatabaseServiceCreatorImplementation(DatabaseServiceCreatorInterface):
    def create_db(self) -> DataBaseProductInterface:
        return PSQLDataBaseProductImplementation()

class MysqlDatabaseServiceCreatorImplementation(DatabaseServiceCreatorInterface):
    def create_db(self) -> DataBaseProductInterface:
        return MysqlDataBaseProductImplementation()

# Client code
def main():
    sqlite_service = SQLITEDatabaseServiceCreatorImplementation()
    sqlite_conn = sqlite_service.create_connection()

    psql_service = PSQLDatabaseServiceCreatorImplementation()
    psql_conn = psql_service.create_connection()

    mysql_service = MysqlDatabaseServiceCreatorImplementation()
    mysql_conn = mysql_service.create_connection()

if __name__ == "__main__":
    main()
```
