this pattern is used for when we have lots of objects and all of them have something in commen but some parts are different we use this .

Here is an awful code that we are going to refactore it

```python
import copy

class Circle:
    def __init__(self, id, radius, color):
        self.id = id
        self.radius = radius
        self.color = color

class Rectangle:
    def __init__(self, id, width, height, color):
        self.id = id
        self.width = width
        self.height = height
        self.color = color

def create_shape(shape_type, id, *args):
    if shape_type == "circle":
        return Circle(id, *args)
    elif shape_type == "rectangle":
        return Rectangle(id, *args)
    else:
        raise ValueError("Unknown shape type")

def clone_shape(shape):
    if isinstance(shape, Circle):
        return copy.deepcopy(shape)
    elif isinstance(shape, Rectangle):
        return copy.deepcopy(shape)
    else:
        raise ValueError("Unknown shape instance")

def main():
    shapes = []

    # Creating shapes
    circle1 = create_shape("circle", "1", 10, "red")
    shapes.append(circle1)
    rectangle1 = create_shape("rectangle", "2", 20, 30, "blue")
    shapes.append(rectangle1)

    # Cloning shapes
    cloned_shapes = []
    for shape in shapes:
        cloned_shapes.append(clone_shape(shape))

    # Display shapes
    for shape in shapes:
        if isinstance(shape, Circle):
            print(f"Circle(ID: {shape.id}, Radius: {shape.radius}, Color: {shape.color})")
        elif isinstance(shape, Rectangle):
            print(f"Rectangle(ID: {shape.id}, Width: {shape.width}, Height: {shape.height}, Color: {shape.color})")

    for shape in cloned_shapes:
        if isinstance(shape, Circle):
            print(f"Cloned Circle(ID: {shape.id}, Radius: {shape.radius}, Color: {shape.color})")
        elif isinstance(shape, Rectangle):
            print(f"Cloned Rectangle(ID: {shape.id}, Width: {shape.width}, Height: {shape.height}, Color: {shape.color})")

if __name__ == "__main__":
    main()
```

now lets refactor it

```python
from copy import deepcopy
from abc import ABC, abstractmethod

class Shape(ABC):
    def __init__(self, id: str, color: str) -> None:
        self.id = id
        self.color = color

    @abstractmethod
    def clone(self) -> 'Shape':
        pass

    @abstractmethod
    def __str__(self) -> str:
        pass


class Circle(Shape):
    def __init__(self, id: str, radius: float, color: str) -> None:
        super().__init__(id, color)
        self.radius = radius

    def clone(self) -> 'Circle':
        return deepcopy(self)

    def __str__(self) -> str:
        return f"Circle(ID: {self.id}, Radius: {self.radius}, Color: {self.color})"


class Rectangle(Shape):
    def __init__(self, id: str, width: float, height: float, color: str) -> None:
        super().__init__(id, color)
        self.width = width
        self.height = height

    def clone(self) -> 'Rectangle':
        return deepcopy(self)

    def __str__(self) -> str:
        return f"Rectangle(ID: {self.id}, Width: {self.width}, Height: {self.height}, Color: {self.color})"


# Client code
def main() -> None:
    c1 = Circle("1", 10, "red")
    c2 = c1.clone()

    print(c1)
    print(c2)

    r1 = Rectangle("2", 20, 30, "blue")
    r2 = r1.clone()

    print(r1)
    print(r2)

if __name__ == "__main__":
    main()
```

so how it is ? we have a interface that has __init__ and the commen attributes between all of the objects .
<br>
then this interface have a clone and display or any other commen fucntions defien in interface
<br>
each object type can be a concrete class and implement __init__ using super() and other attributes / implement clone by deepcopy or shallowcopy and then the other functions as they want


here another example


```python
class Config:
    def __init__(self, environment):
        self.environment = environment
        if environment == "development":
            self.database_url = "sqlite:///dev.db"
            self.debug = True
            self.secret_key = "dev-secret"
        elif environment == "staging":
            self.database_url = "sqlite:///staging.db"
            self.debug = True
            self.secret_key = "staging-secret"
        elif environment == "production":
            self.database_url = "sqlite:///prod.db"
            self.debug = False
            self.secret_key = "prod-secret"
        else:
            raise ValueError("Unknown environment")

    def clone(self):
        if self.environment == "development":
            return Config("development")
        elif self.environment == "staging":
            return Config("staging")
        elif self.environment == "production":
            return Config("production")
        else:
            raise ValueError("Unknown environment")

    def __str__(self):
        return f"Config(Environment: {self.environment}, Database URL: {self.database_url}, Debug: {self.debug}, Secret Key: {self.secret_key})"


def main():
    dev_config = Config("development")
    staging_config = Config("staging")
    prod_config = Config("production")

    # Clone configurations
    cloned_dev_config = dev_config.clone()
    cloned_staging_config = staging_config.clone()
    cloned_prod_config = prod_config.clone()

    # Display original and cloned configurations
    print("Original configurations:")
    print(dev_config)
    print(staging_config)
    print(prod_config)

    print("\nCloned configurations:")
    print(cloned_dev_config)
    print(cloned_staging_config)
    print(cloned_prod_config)

if __name__ == "__main__":
    main()
```

```python
from abc import ABC ,abstractmethod
from copy import deepcopy
class Config(ABC):
  def __init__(self,database_url,debug,secret_key):
    self.database_url =database_url
    self.debug = debug
    self.secret_key = secret_key

  @abstractmethod
  def clone(self):
    pass
  @abstractmethod
  def __str__(self):
    pass


class Development(Config):
  def __init__(self,database_url,debug,secret_key):
    super().__init__(database_url = database_url,debug = debug,secret_key = secret_key)

  def clone(self):
    return deepcopy(self)

  def __str__(self):
      return f"Config(Environment: {self.__class__.__name__}, Database URL: {self.database_url}, Debug: {self.debug}, Secret Key: {self.secret_key})"


class Staging(Config):
  def __init__(self,database_url,debug,secret_key):
    super().__init__(database_url = database_url,debug = debug,secret_key = secret_key)

  def clone(self):
    return deepcopy(self)

  def __str__(self):
      return f"Config(Environment: {self.__class__.__name__}, Database URL: {self.database_url}, Debug: {self.debug}, Secret Key: {self.secret_key})"



class Production(Config):
  def __init__(self,database_url,debug,secret_key):
    super().__init__(database_url = database_url,debug = debug,secret_key = secret_key)

  def clone(self):
    return deepcopy(self)

  def __str__(self):
      return f"Config(Environment: {self.__class__.__name__}, Database URL: {self.database_url}, Debug: {self.debug}, Secret Key: {self.secret_key})"



a=Development("a","a","a")
b=a.clone()

c=Staging("b","b","b")
d=c.clone()

e=Production("c","c","c")
f=e.clone()

print(a)
print(b)
print(c)
print(d)
print(e)
print(f)


```
