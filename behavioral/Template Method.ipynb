```python
# refactored code
from abc import ABC , abstractmethod
class BeverageMaker(ABC):

  @abstractmethod
  def brew(self):
    pass


  @abstractmethod
  def pour_in_cup(self):
    pass


  def boil_water(self):
    print("Boiling water")


  def add_tastes(self,tastes):
    for taste in tastes:
      print(f"Adding {taste}")


  def make_beverage(self,tastes):
    self.boil_water()
    self.brew()
    self.pour_in_cup()
    self.add_tastes(tastes)

class CoffeeMaker(BeverageMaker):

  def brew(self):
    print("Brewing coffee")
  def pour_in_cup(self):
    print("Pouring coffee into the cup")


class TeaMaker(BeverageMaker):

  def brew(self):
    print("Steeping tea")
  def pour_in_cup(self):
    print("Pouring tea into the cup")


coffee=CoffeeMaker()
coffee.make_beverage(["sugar","milk"])
tea=TeaMaker()
tea.make_beverage(["lemon"])


```
