**Iterator is one of the behavioral design patterns that used for traverse a big list or any type of iterables in sequesntial order
<br>
it has two type of implemetation :<br>
1- implementing _ _ next _ _ , _ _ iter _ _ <br>
2- implementing next , has_next<br>


```python
# way 1
class MyIterator:
  def __init__(self,data):
    self.data=data
    self.index=0

  def __iter__(self):
    return self

  def __next__(self):
    if self.index>=len(self.data):
      print("no more elements")
      raise StopIteration

    value=self.data[self.index]
    self.index+=1
    return value

a=[1,2,3,4,5,6,7,8,9,0]
iterator=MyIterator(a)
print(a)
print(iterator.__next__())
print(iterator.__next__())
print(iterator.__next__())
print(iterator.__next__())
print(iterator.__next__())
print(iterator.__next__())
print(iterator.__next__())
print(iterator.__next__())
print(iterator.__next__())
print(iterator.__next__())
print(iterator.__next__())
```

```python
# way 2
class MyIterator2:
  def __init__(self,data):
    self.data=data
    self.index=0
  def has_next(self):
    if self.index>=len(self.data):
      return False
    return True
  def next(self):
    if not self.has_next():
      print("no more elements")
      raise StopIteration
    value=self.data[self.index]
    self.index+=1
    return value

a=[1,2,3,4,5,6,7,8,9,0]
iterator=MyIterator2(a)
print(a)
print(iterator.next())
print(iterator.next())
print(iterator.next())
print(iterator.next())
print(iterator.next())
print(iterator.next())
print(iterator.next())
print(iterator.next())
print(iterator.next())
print(iterator.next())
print(iterator.next())
```
