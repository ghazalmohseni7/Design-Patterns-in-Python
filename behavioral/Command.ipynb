```python
#refactored
from abc import ABC, abstractmethod

class Commands(ABC):
    @abstractmethod
    def execute(self):
        pass

class LightReceiver:
    def on(self):
        print("Light is ON")

    def off(self):
        print("Light is OFF")

class LightOnCommand(Commands):
    def __init__(self, light):
        self.light = light

    def execute(self):
        self.light.on()

class LightOffCommand(Commands):
    def __init__(self, light):
        self.light = light

    def execute(self):
        self.light.off()

class RemoteControlInvoker:
    def __init__(self):
        self.command_on = None
        self.command_off = None

    def set_command(self, on_command, off_command):
        self.command_on = on_command
        self.command_off = off_command

    def press_on(self):
        if self.command_on:
            self.command_on.execute()

    def press_off(self):
        if self.command_off:
            self.command_off.execute()

# Create a Light object
light = LightReceiver()

# Create Command objects for turning the light on and off
light_on = LightOnCommand(light)
light_off = LightOffCommand(light)

# Create a RemoteControl object
remote = RemoteControlInvoker()

# Set the commands on the remote
remote.set_command(light_on, light_off)

# Use the remote to turn the light on and off
remote.press_on()  # Output: Light is ON
remote.press_off()  # Output: Light is OFF

```

```python
# REFACTORED
from abc import ABC, abstractmethod

# Command Interface
class Command(ABC):
    @abstractmethod
    def execute(self):
        pass

    @abstractmethod
    def undo(self):
        pass

# Receiver
class TextReceiver:
    def __init__(self):
        self.text = ""
        self.history = []

    def type_text(self, text):
        self.history.append(self.text)
        self.text += text
        print(f"Text after typing: {self.text}")

    def delete_text(self, num_chars):
        self.history.append(self.text)
        self.text = self.text[:-num_chars]
        print(f"Text after deletion: {self.text}")

    def undo(self):
        if self.history:
            self.text = self.history.pop()
            print(f"Text after undo: {self.text}")
        else:
            print("Nothing to undo.")

# Concrete Commands
class TypeCommand(Command):
    def __init__(self, text: str, receiver: TextReceiver):
        self.text = text
        self.receiver = receiver

    def execute(self):
        self.receiver.type_text(self.text)

    def undo(self):
        self.receiver.undo()

class DeleteCommand(Command):
    def __init__(self, num_chars: int, receiver: TextReceiver):
        self.num_chars = num_chars
        self.receiver = receiver

    def execute(self):
        self.receiver.delete_text(self.num_chars)

    def undo(self):
        self.receiver.undo()

# Invoker
class TyperInvoker:
    def __init__(self):
        self.command = None

    def set_command(self, command: Command):
        self.command = command

    def execute(self):
        if self.command:
            self.command.execute()

    def undo(self):
        if self.command:
            self.command.undo()

# Example Usage
text_receiver = TextReceiver()

# Create Command objects
type_command = TypeCommand("Hello, World!", text_receiver)
delete_command = DeleteCommand(6, text_receiver)  # Delete last 6 characters

# Create the Invoker
invoker = TyperInvoker()

# Set and execute the type command
invoker.set_command(type_command)
invoker.execute()  # Output: Text after typing: Hello, World!

# Set and execute the delete command
invoker.set_command(delete_command)
invoker.execute()  # Output: Text after deletion: Hello,

# Undo the last delete command
invoker.undo()  # Output: Text after undo: Hello, World!

# Undo the last type command
invoker.set_command(type_command)
invoker.undo()  # Output: Text after undo: (empty)
```

```python
# SECOND REFACTOR
from abc import ABC, abstractmethod

class Command(ABC):
    @abstractmethod
    def execute(self):
        pass

    @abstractmethod
    def undo(self):
        pass

class SmartLight:
    def turn_on(self):
        print("Smart Light is ON")

    def turn_off(self):
        print("Smart Light is OFF")

class SmartThermostat:
    def __init__(self):
        self.temperature = None

    def set_temperature(self, temperature):
        self.temperature = temperature
        print(f"Smart Thermostat set to {self.temperature}°C")

class CommandLight(Command):
    def __init__(self, light: SmartLight):
        self.light = light

    def execute(self):
        self.light.turn_on()

    def undo(self):
        self.light.turn_off()

class CommandTemp(Command):
    def __init__(self, thermostat: SmartThermostat, temp: int):
        self.thermostat = thermostat
        self.temp = temp
        self.prev_temp = None

    def execute(self):
        self.prev_temp = self.thermostat.temperature
        self.thermostat.set_temperature(self.temp)

    def undo(self):
        if self.prev_temp is not None:
            self.thermostat.set_temperature(self.prev_temp)
            print(f"Reverted temperature to {self.prev_temp}°C")

class SmartHomeController:
    def __init__(self):
        self.command = None

    def set_command(self, command: Command):
        self.command = command

    def execute_command(self):
        if self.command:
            self.command.execute()

    def undo_command(self):
        if self.command:
            self.command.undo()

# Example Usage
light = SmartLight()
thermostat = SmartThermostat()

# Create Command objects
light_on_command = CommandLight(light)
temp_command = CommandTemp(thermostat, 22)

# Create the Invoker
controller = SmartHomeController()

# Set and execute light command
controller.set_command(light_on_command)
controller.execute_command()  # Output: Smart Light is ON
controller.undo_command()     # Output: Smart Light is OFF

# Set and execute temperature command
controller.set_command(temp_command)
controller.execute_command()  # Output: Smart Thermostat set to 22°C
controller.undo_command()     # Output: Reverted temperature to previous setting
```
